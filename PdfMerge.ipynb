```python
from PyPDF2 import PdfFileMerger

import os
source_dir = '.\\Input\\'
filename = input("Enter the file name: ")

merger = PdfFileMerger(strict=False)
#source_dir = os.getcwd()

files = os.listdir(source_dir)
for item in files:
    if item.endswith("pdf"):
        sou = str(source_dir+item)
        merger.append(sou)  
        
merger.write(filename+'.pdf')
merger.close()
    
```

```python
from PyPDF2 import PdfWriter, PdfReader, PdfFileMerger

import os
source_dir = '.\\Input\\'
destination = '.\\Output\\'

files = os.listdir(source_dir)
for file in files:
    if file.endswith('.pdf'):
        break
        
with open(source_dir+file, "rb") as og_file:
    Reader = PdfReader(og_file)
    num_pages = len(Reader.pages)
    for i in range(Reader.numPages):
        pdf_writer = PdfWriter()
        pdf_writer.add_page(Reader.pages[i])
        with open(destination+"Page"+str(i)+".pdf", "wb") as f:
            pdf_writer.write(f)

```

Rotate PDF pages


```python
from PyPDF2 import PdfFileReader, PdfFileWriter
filename = input("Enter the file name: ")
print(filename)
rotation = int(input("Enter the number of degrees to rorate by :"))
with open(filename, 'rb') as f:
    Reader = PdfFileReader(f)
    writer = PdfFileWriter()
    for pagenum in range(Reader.numPages):
        page = Reader.getPage(pagenum)
        page.rotateClockwise(rotation)
        writer.addPage(page)
    NewName = ("Rotated"+str(rotation)+filename)
    with open(NewName, 'wb') as Newf:
        writer.write(Newf)
```
